```
import pywedge as pw
```

```
import pandas as pd
train = pd.read_csv('https://raw.githubusercontent.com/taknev83/datasets/master/HR_Employee_Attrition.csv')
```

Instantiate the Pywedge_Charts class.

The Pywedge_Charts class takes following inputs,

Inputs:
1. Dataframe
2. c = any redundant column to be removed (like ID column etc., at present supports a single column removal, subsequent version will provision multiple column removal requirements)
3. y = target column name as a string 
        
Returns:

 Charts widget

```
mc = pw.Pywedge_Charts(train, c=None, y = 'Attrition')
```

Call make_charts method from the instantiated class

```
charts = mc.make_charts()
```
